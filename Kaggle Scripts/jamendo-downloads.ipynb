# **Pop 1000+**

```python
import requests
songs = requests.get('https://api.jamendo.com/v3.0/tracks/?client_id=df9041d4&format=jsonpretty&durationbetween=120_240&tags=pop&include=musicinfo&boost=popularity_total&limit=100&offset=1400').json()['results']
print("Songs received")
# for song in songs:
#     print(song['audiodownload'])
rock = 0
nd = 0
s = 0
for x in range (0, len(songs)):
    print(f"Processing song {x}")
    song = songs[x]
    if 'rock' in song['musicinfo']['tags']['genres']:
        rock +=1
        continue
    try:
        songlink = requests.get(song['audiodownload'])
        title = song['name'] + '.mp3'
        open(title, 'wb').write(songlink.content)
        s += 1
    except:
        nd += 1
        continue
print("Rock: ", rock)
print("Download failed: ", nd)
print("Successful: ", s)
```
